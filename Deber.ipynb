# Deber. Series de Tylor y Polinomios de Lagrange
## Determine el orden de la mejor aproximación para las siguientes funciones, usando la Serie de Taylor y el Polinomio de Lagrange:
### 1. 1/25*x^2 + 1, x_o=0 

```python
import numpy as np
# Definir serie de tylor de la funcion 
#definir primero la funcion inicial:
def f(x):
    return 1/(25*x**2 +1) 
def funtylor0(x):
    return np.ones_like(x) #dado que retornar 1 es diferente en el lenguaje, se lo define de esta manera para que devuelva un arreglo
def funtylor2(x):
    return 1-25*x**2
def funtylor4(x):
    return 1-25*x**2+625*x**4
def funtylor6(x):
    return 1-25*x**2+625*x**4+15625*x**6

#para el polinomio se uso los puntos -0.1,0,0.1 dado que fuera de ello diverge la funcion
def lagrange_fun1(x):
    return 1 - 20 * x**2

#Graficar la serie
import matplotlib.pyplot as plt
x = np.linspace(-0.3, 0.3, 500)
plt.figure(figsize=(10, 6))

plt.plot(x, f(x), label='Función Original $f(x)$', color='black', linewidth=2)
plt.plot(x, funtylor0(x), label='Taylor Orden 0', linestyle='--')
plt.plot(x, funtylor2(x), label='Taylor Orden 2', linestyle='--')
plt.plot(x, funtylor4(x), label='Taylor Orden 4', linestyle='--')
plt.plot(x, funtylor6(x), label='Taylor Orden 6', linestyle='--')
#polinomio de lagrange
plt.plot(x, lagrange_fun1(x), label='Lagrange Orden 2', linestyle='-.', color='purple')


plt.title('Aproximación de la Serie de Taylor', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linewidth=0.5, linestyle=':')
plt.axvline(0, color='gray', linewidth=0.5, linestyle=':')
plt.ylim(-0.5, 1.5)
plt.grid(True, alpha=0.3)
plt.show()
```

### 2. arctan(x), x_o=0 

```python
#aplicando el mismo concepto anterior, es decir usando la funcion geométrica 1/(1-x)=1+x^2+x^3+...
def f2(x):
    return np.arctan(x)

def f2tylor1(x):
    return x
def f2tylor3(x):
    return x-(x**3/3)
def f2tylor5(x):
    return x-(x**3/3)+(x**5/5)
def f2tylor7(x):
    return x-(x**3/3)+(x**5/5)-(x**7/7)
#se usó los puntos -0.5,0,0.5 dado que fuera de este rango por [-1,1] la funcion diverge
def lagrange_arctan(x):
    return 0.9272 * x

x = np.linspace(-1.5, 1.5, 500)

# 4. Graficamos
plt.figure(figsize=(10, 6))

plt.plot(x, f2(x), label='Función Original $\\arctan(x)$', color='black', linewidth=2)
plt.plot(x, f2tylor1(x), label='Taylor Orden 1', linestyle='--')
plt.plot(x, f2tylor3(x), label='Taylor Orden 3', linestyle='--')
plt.plot(x, f2tylor5(x), label='Taylor Orden 5', linestyle='--')
plt.plot(x, f2tylor7(x), label='Taylor Orden 7', linestyle='--')
#polinomio de lagrange
plt.plot(x, lagrange_arctan(x), label='Lagrange Orden 1', linestyle='-.', color='purple')

plt.title('Aproximación de la Serie de Taylor para $\\arctan(x)$', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linewidth=0.5, linestyle=':')
plt.axvline(0, color='gray', linewidth=0.5, linestyle=':')
plt.ylim(-2, 2) # Limitamos Y para observar la divergencia fuera de [-1, 1]
plt.grid(True, alpha=0.3)
plt.show()
```
