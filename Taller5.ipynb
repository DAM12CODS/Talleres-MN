```python
xs4 = [
    3.38,
    0.35,
    2.07,
    -0.45,
    -0.56,
    -1.06,
    -2.78,
    3.08,
    -3.99,
    -5.00,
    -3.18,
    -1.97,
    2.37,
    -1.67,
    4.09,
    -4.60,
    2.68,
    2.78,
    -3.79,
    -1.87
]

ys4 = [
    15.65,
    -11.20,
    -4.00,
    18.03,
    7.94,
    15.32,
    -4.40,
    1.39,
    -11.92,
    -90.24,
    6.92,
    28.35,
    -2.41,
    -5.47,
    35.91,
    -55.53,
    0.77,
    4.79,
    -27.05,
    2.85
]
sumXs4 = sum(xs4)
sumYs4 = sum(ys4)
sumXs4AlCuadrado = sum(x**2 for x in xs4)
sumYs4AlCubo = sum(y**3 for y in ys4)
sumXs4AlCubo = sum(x**3 for x in xs4)
print("ΣXs4 =", sumXs4)
print("ΣYs4 =", sumYs4)
print("ΣXs4² =", sumXs4AlCuadrado)
print("ΣYs4³ =", sumYs4AlCubo)


## sistema resultante, al ser cuadrática tengo que tener un sistema de cuatro ecuaciones: 
a=1.0450
b=0.0313
c=-8.8807
d=2.7623
def nuevFu(x): 
    return a*x**3 + b*x**2 + c*x + d 

plt.figure(figsize=(8, 6))
plt.scatter(xs4, ys4, color='purple', marker='o', label='xs4 vs ys4')
plt.xlabel('xs4')
plt.ylabel('ys4')
plt.title('Puntos xs4 y ys4')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x_plot4 = np.linspace(min(xs4) - 1, max(xs4) + 1, 200)
y_nuevFu = [nuevFu(x) for x in x_plot4]

plt.figure(figsize=(8, 6))
plt.scatter(xs4, ys4, color='purple', marker='o', label='Puntos xs4, ys4')
plt.plot(x_plot4, y_nuevFu, color='red', label='nuevFu(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Puntos xs4, ys4 y función nuevFu')
plt.grid(True)
plt.show()
```

```python

#Conjunto de datos 2
x_datos = np.array([3.38, 0.35, 2.07, -0.45, -0.56, -1.06, -2.78, 3.08, -3.99, -5.00, 
                    -3.18, -1.97, 2.37, -1.67, 4.09, -4.60, 2.68, 2.78, -3.79, -1.87])
y_datos = np.array([15.65, -11.20, -4.00, 18.03, 7.94, 15.32, -4.40, 1.39, -11.92, -90.24, 
                    6.92, 28.35, -2.41, -5.47, 35.91, -55.53, 0.77, 4.79, -27.05, 2.85])


# 3. Cálculo de predicciones y errores
y_predicho = nuevFu(x_datos)
errores_absolutos = np.abs(y_datos - y_predicho)

# Estadísticas
error_promedio = np.mean(errores_absolutos)
desviacion_error = np.std(errores_absolutos)

# Encontrar índices de error máximo y mínimo
idx_max = np.argmax(errores_absolutos)
idx_min = np.argmin(errores_absolutos)

# Mostrar resultados en consola
print("--- ESTADÍSTICAS DE ERROR ---")
print(f"Error Absoluto Promedio: {error_promedio:.4f}")
print(f"Desviación Estándar del Error: {desviacion_error:.4f}")
print(f"\nError Máximo: {errores_absolutos[idx_max]:.4f} en el punto ({x_datos[idx_max]}, {y_datos[idx_max]})")
print(f"Error Mínimo: {errores_absolutos[idx_min]:.4f} en el punto ({x_datos[idx_min]}, {y_datos[idx_min]})")

# 4. Gráfica
plt.figure(figsize=(10, 6))

# Dibujar los puntos de datos y la curva de ajuste
plt.scatter(x_datos, y_datos, color='gray', alpha=0.7, label='Datos Originales')
x_curva = np.linspace(min(x_datos) - 0.5, max(x_datos) + 0.5, 500)
plt.plot(x_curva, nuevFu(x_curva), color='blue', label='Ajuste Mínimos Cuadrados')

# Destacar Punto de Error Máximo (Rojo)
plt.scatter(x_datos[idx_max], y_datos[idx_max], color='red', s=120, edgecolors='black', zorder=5,
            label=f'Error Máx ({errores_absolutos[idx_max]:.2f})')
plt.annotate('Error Máx', (x_datos[idx_max], y_datos[idx_max]), textcoords="offset points", 
             xytext=(10,-10), ha='left', color='red', weight='bold')

# Destacar Punto de Error Mínimo (Verde)
plt.scatter(x_datos[idx_min], y_datos[idx_min], color='green', s=120, edgecolors='black', zorder=5,
            label=f'Error Mín ({errores_absolutos[idx_min]:.2f})')
plt.annotate('Error Mín', (x_datos[idx_min], y_datos[idx_min]), textcoords="offset points", 
             xytext=(10,10), ha='left', color='green', weight='bold')

# Personalización del gráfico
plt.title('Ajuste Cúbico con Puntos de Error Extremos')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
