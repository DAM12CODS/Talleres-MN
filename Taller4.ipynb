# Taller 4 - Splines Cúbicos

```python
import numpy as np
import matplotlib.pyplot as plt


## Definicion de los polinomios ya calculados para cada ejercicio

def spline_ejercicio1(x):
    #(0,1), (1,5), (2,3)
    if 0 <= x <= 1:
        return 1 + 5.5 * x - 1.5 * x**3
    elif 1 < x <= 2:
        return 5 + 1 * (x - 1) - 4.5 * (x - 1)**2 + 1.5 * (x - 1)**3
    return None

def spline_ejercicio2(x):
    #(0,-5), (1,-4), (2,3)
    if 0 <= x <= 1:
        return -5 - 0.5 * x + 1.5 * x**3
    elif 1 < x <= 2:
        return -4 + 4 * (x - 1) + 4.5 * (x - 1)**2 - 1.5 * (x - 1)**3
    return None


def spline_ejercicio3(x):
    #(0,-1), (1,1), (2,5), (3,2)
    if 0 <= x <= 1:
        return -1 + 1 * x + 1 * x**3
    elif 1 < x <= 2:
        return 1 + 4 * (x - 1) + 3 * (x - 1)**2 - 3 * (x - 1)**3
    elif 2 < x <= 3:
        return 5 + 1 * (x - 2) - 6 * (x - 2)**2 + 2 * (x - 2)**3
    return None

# Vectorizar las funciones para que puedan evaluar arreglos de NumPy de golpe
spline1_vec = np.vectorize(spline_ejercicio1)
spline2_vec = np.vectorize(spline_ejercicio2)
spline3_vec = np.vectorize(spline_ejercicio3)


plt.figure(figsize=(15, 5))

# --- Gráfico Ejercicio 1 ---
plt.subplot(1, 3, 1)
x_val1 = np.linspace(0, 2, 200)
plt.plot(x_val1, spline1_vec(x_val1), color='blue', linewidth=2, label='Spline $S(x)$')
plt.scatter([0, 1, 2], [-5, -4, 3], color='red', zorder=5, label='Nodos')
plt.title('Ejercicio 1\n(0,-5), (1,-4), (2,3)')
plt.grid(True, linestyle='--')


# --- Gráfico Ejercicio 2 ---
plt.subplot(1, 3, 2)
x_val2 = np.linspace(0, 2, 200)
plt.plot(x_val2, spline2_vec(x_val2), color='green', linewidth=2, label='Spline $S(x)$')
plt.scatter([0, 1, 2], [1, 5, 3], color='red', zorder=5, label='Nodos')
plt.title('Ejercicio 2\n(0,1), (1,5), (2,3)')
plt.grid(True, linestyle='--')


# --- Gráfico Ejercicio 3 ---
plt.subplot(1, 3, 3)
x_val3 = np.linspace(0, 3, 300)
plt.plot(x_val3, spline3_vec(x_val3), color='purple', linewidth=2, label='Spline $S(x)$')
plt.scatter([0, 1, 2, 3], [-1, 1, 5, 2], color='red', zorder=5, label='Nodos')
plt.title('Ejercicio 3: 3 Tramos\n(0,-1), (1,1), (2,5), (3,2)')
plt.grid(True, linestyle='--')


plt.tight_layout()
plt.show()
```
