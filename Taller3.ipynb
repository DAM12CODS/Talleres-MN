```python
import numpy as np
import matplotlib.pyplot as plt
import math
# Definir la función coseno
def coseno(x):
    return np.cos(x) ## es la función coseno

# Dominio de visualización
x = np.linspace(-2, 2, 1000)
y = coseno(x)

plt.figure(figsize=(8,4))
plt.plot(x, y, label='cos(x)', color='blue')

plt.title('Función Coseno')
plt.xlabel('x')
plt.ylabel('cos(x)')

plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid(True)

# Marcas en el eje x
xticks = [
    -np.pi/2,
    -np.pi/4,
    0,
    np.pi/4,
    np.pi/2
]

xticklabels = [
    r'$-\frac{\pi}{2}$',
    r'$-\frac{\pi}{4}$',
    r'$0$',
    r'$\frac{\pi}{4}$',
    r'$\frac{\pi}{2}$'
]

plt.xticks(xticks, xticklabels)

# Limitar la vista entre -2 y 2
plt.xlim(-4, 4)

plt.legend()
plt.show()


### Aproximaciones de Taylor para coseno
def taylor_coseno(x, n):
    """Calcula la aproximación de Taylor de coseno hasta el orden n."""
    approximation = 0
    for k in range(n + 1):
        approximation += ((-1)**k * x**(2 * k)) / math.factorial(2 * k)# Esta s la formula de aproximacion de taylor para el coseno
    return approximation
# Graficar la función coseno y sus aproximaciones de Taylor
x = np.linspace(-2, 2, 1000)
y_coseno = coseno(x)
y_taylor_0 = taylor_coseno(x, n=0)  # Aproximación de orden 0
y_taylor_1 = taylor_coseno(x, n=1)  # Aproximación de orden 1
y_taylor_2 = taylor_coseno(x, n=2)  # Aproximación de orden 2
y_taylor_3 = taylor_coseno(x, n=3)  # Aproximación de orden 3
y_taylor_4 = taylor_coseno(x, n=4)  # Aproximación de orden 4
plt.figure(figsize=(10, 6))
plt.plot(x, y_coseno, label='cos(x)', color='blue')
#plt.plot(x, y_taylor_0, label='Taylor orden 0', linestyle='--')
#plt.plot(x, y_taylor_1, label='Taylor orden 1', linestyle='--')
plt.plot(x, y_taylor_2, label='Taylor orden 2', linestyle='--')
#plt.plot(x, y_taylor_3, label='Taylor orden 3', linestyle='--')
#plt.plot(x, y_taylor_4, label='Taylor orden 4', linestyle='--')
plt.title('Función Coseno y Aproximaciones de Taylor')  



### Primera ecuacion de la aproximacion de taylor para coseno con x_0=1
def taylor_cos_1(x): 
    return 0.54
#Graficar la primera aproximacion de taylor para coseno con x_0=1
def taylor_cos_2(x): 
    return 0.54-0.84*x
def taylor_cos_3(x): 
    return 0.54-0.84*x-(0.54/2)*x**2
def taylor_cos_4(x): 
    return 0.54-0.84*x-(0.54/2)*x**2+(0.84/6)*x**3+(0.54/24)*x**4

```

```python

def taylor_cos_g0(x): 
    return np.cos(1) * np.ones_like(x)

def taylor_cos_g1(x): 
    return np.cos(1) - np.sin(1)*(x - 1)

def taylor_cos_g2(x): 
    return np.cos(1) - np.sin(1)*(x - 1) - (np.cos(1)/2)*(x - 1)**2

def taylor_cos_g3(x): 
    return np.cos(1) - np.sin(1)*(x - 1) - (np.cos(1)/2)*(x - 1)**2 + (np.sin(1)/6)*(x - 1)**3

def taylor_cos_g4(x): 
    return np.cos(1) - np.sin(1)*(x - 1) - (np.cos(1)/2)*(x - 1)**2 + (np.sin(1)/6)*(x - 1)**3 + (np.cos(1)/24)*(x - 1)**4

# Configuración del eje X
x = np.linspace(-2, 4, 1000) 
y_coseno = np.cos(x)          

# Evaluamos pasando 'x' directamente ya que la resta (x-1) se hace internamente
y_taylor_0 = taylor_cos_g0(x)
y_taylor_1 = taylor_cos_g1(x)
y_taylor_2 = taylor_cos_g2(x)
y_taylor_3 = taylor_cos_g3(x)
y_taylor_4 = taylor_cos_g4(x)

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y_coseno, label='cos(x)', color='blue', linewidth=2)
plt.plot(x, y_taylor_0, label='Taylor Grado 0 (Constante)', linestyle='--')
plt.plot(x, y_taylor_1, label='Taylor Grado 1 (Lineal)', linestyle='--')
plt.plot(x, y_taylor_2, label='Taylor Grado 2 (Cuadrático)', linestyle='--')
plt.plot(x, y_taylor_3, label='Taylor Grado 3 (Cúbico)', linestyle='--')
plt.plot(x, y_taylor_4, label='Taylor Grado 4 (Cuártico)', linestyle='--')



plt.title('Función Coseno y Aproximaciones de Taylor en x0 = 1')
plt.xlabel('x') 
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.ylim(-2, 2)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

## Grafica de las aproximaciones de tylor de grado 4 de la siguiente función : 1 / 1-x

```python
## Primer derivamos la funcion: 1/(1-x)
def f_taylor0(x):
    return 1/(1-x)  
def df_taylor1(x):
    return (1/(1-x)**2)*x
def d2f_taylor2(x):
    return (2/(1-x)**3)*x**2/2
def d3f_taylor3(x):
    return (6/(1-x)**4)*x**3/6
def d4f_taylor4(x):
    return (24/(1-x)**5)*x**4/24


## Graficar la funcion 1/(1-x) y sus aproximaciones de Taylor
x = np.linspace(-2, 0.9, 1000)  # Evitamos x=1 para evitar la singularidad
y_funcion = f_taylor0(x)    
y_taylor_0 = f_taylor0(x)  # Aproximación de orden 0
y_taylor_1 = f_taylor0(x) + df_taylor1(x)  # Aproximación de orden 1
y_taylor_2 = f_taylor0(x) + df_taylor1(x) + d2f_taylor2(x)  # Aproximación de orden 2
y_taylor_3 = f_taylor0(x) + df_taylor1(x) + d2f_taylor2(x) + d3f_taylor3(x)  # Aproximación de orden 3
y_taylor_4 = f_taylor0(x) + df_taylor1(x) + d2f_taylor2(x) + d3f_taylor3(x) + d4f_taylor4(x)  # Aproximación de orden 4 
plt.figure(figsize=(10, 6))
plt.plot(x, y_funcion, label='f(x) = 1/(1-x)', color='blue')
plt.plot(x, y_taylor_0, label='Taylor orden 0', linestyle='--')
plt.plot(x, y_taylor_1, label='Taylor orden 1', linestyle='--')
plt.plot(x, y_taylor_2, label='Taylor orden 2', linestyle='--')
plt.plot(x, y_taylor_3, label='Taylor orden 3', linestyle='--')
plt.plot(x, y_taylor_4, label='Taylor orden 4', linestyle='--')
plt.title('Función 1/(1-x) y Aproximaciones de Taylor')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.xlim(-1, 0.9)
plt.ylim(-5, 5)
plt.grid(True)
plt.show()
```

## Grafica de las aproximaciones de tylor de grado 4 de la siguiente función : ln(x), x_0 = 1

```python
# Ahora para la función ln(x) con x_0=1
def f_ln_tylor0(x):
    return np.log(x)
def df_ln_tylor1(x):
    return (1/(x-1) )*(x-1)
def d2f_ln_tylor2(x):
    return -(x-1)**2/2*x**2
def d3f_ln_tylor3(x):
    return 2*(x-1)**3/6*x**3
def d4f_ln_tylor4(x):
    return -6*(x-1)**4/24*x**4  

## Graficar la función ln(x) y sus aproximaciones de Taylor
x = np.linspace(0.1, 3, 1000)  
y_funcion = f_ln_tylor0(x)
y_taylor_0 = f_ln_tylor0(x) # Aproximación de orden 0
y_taylor_1 = f_ln_tylor0(x) + df_ln_tylor1(x)  # Aproximación de orden 1
y_taylor_2 = f_ln_tylor0(x) + df_ln_tylor1(x) + d2f_ln_tylor2(x)  # Aproximación de orden 2
y_taylor_3 = f_ln_tylor0(x) + df_ln_tylor1(x) + d2f_ln_tylor2(x) + d3f_ln_tylor3(x)  # Aproximación de orden 3
y_taylor_4 = f_ln_tylor0(x) + df_ln_tylor1(x) + d2f_ln_tylor2(x) + d3f_ln_tylor3(x) + d4f_ln_tylor4(x)  # Aproximación de orden 4
plt.figure(figsize=(10, 6))
plt.plot(x, y_funcion, label='f(x) = ln(x)', color='blue')
plt.plot(x, y_taylor_1, label='Taylor orden 1', linestyle='--')
plt.plot(x, y_taylor_2, label='Taylor orden 2', linestyle='--')
plt.plot(x, y_taylor_3, label='Taylor orden 3', linestyle='--')
plt.plot(x, y_taylor_4, label='Taylor orden 4', linestyle='--')
plt.title('Función ln(x) y Aproximaciones de Taylor')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.xlim(0, 3)
plt.ylim(-2, 2)
plt.grid(True)
plt.show()
```
