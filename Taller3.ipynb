## Grafica de las aproximaciones de tylor de grado 4 de la siguiente función : 1 / 1-x

```python
## Primer derivamos la funcion: 1/(1-x)
def f_taylor0(x):
    return 1/(1-x)  
def df_taylor1(x):
    return (1/(1-x)**2)*x
def d2f_taylor2(x):
    return (2/(1-x)**3)*x**2/2
def d3f_taylor3(x):
    return (6/(1-x)**4)*x**3/6
def d4f_taylor4(x):
    return (24/(1-x)**5)*x**4/24


## Graficar la funcion 1/(1-x) y sus aproximaciones de Taylor
x = np.linspace(-2, 0.9, 1000)  # Evitamos x=1 para evitar la singularidad
y_funcion = f_taylor0(x)    
y_taylor_0 = f_taylor0(x)  # Aproximación de orden 0
y_taylor_1 = f_taylor0(x) + df_taylor1(x)  # Aproximación de orden 1
y_taylor_2 = f_taylor0(x) + df_taylor1(x) + d2f_taylor2(x)  # Aproximación de orden 2
y_taylor_3 = f_taylor0(x) + df_taylor1(x) + d2f_taylor2(x) + d3f_taylor3(x)  # Aproximación de orden 3
y_taylor_4 = f_taylor0(x) + df_taylor1(x) + d2f_taylor2(x) + d3f_taylor3(x) + d4f_taylor4(x)  # Aproximación de orden 4 
plt.figure(figsize=(10, 6))
plt.plot(x, y_funcion, label='f(x) = 1/(1-x)', color='blue')
plt.plot(x, y_taylor_0, label='Taylor orden 0', linestyle='--')
plt.plot(x, y_taylor_1, label='Taylor orden 1', linestyle='--')
plt.plot(x, y_taylor_2, label='Taylor orden 2', linestyle='--')
plt.plot(x, y_taylor_3, label='Taylor orden 3', linestyle='--')
plt.plot(x, y_taylor_4, label='Taylor orden 4', linestyle='--')
plt.title('Función 1/(1-x) y Aproximaciones de Taylor')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.xlim(-1, 0.9)
plt.ylim(-5, 5)
plt.grid(True)
plt.show()
```

## Grafica de las aproximaciones de tylor de grado 4 de la siguiente función : ln(x), x_0 = 1

```python
# Ahora para la función ln(x) con x_0=1
def f_ln_tylor0(x):
    return np.log(x)
def df_ln_tylor1(x):
    return (1/(x-1) )*(x-1)
def d2f_ln_tylor2(x):
    return -(x-1)**2/2*x**2
def d3f_ln_tylor3(x):
    return 2*(x-1)**3/6*x**3
def d4f_ln_tylor4(x):
    return -6*(x-1)**4/24*x**4  

## Graficar la función ln(x) y sus aproximaciones de Taylor
x = np.linspace(0.1, 3, 1000)  
y_funcion = f_ln_tylor0(x)
y_taylor_0 = f_ln_tylor0(x) # Aproximación de orden 0
y_taylor_1 = f_ln_tylor0(x) + df_ln_tylor1(x)  # Aproximación de orden 1
y_taylor_2 = f_ln_tylor0(x) + df_ln_tylor1(x) + d2f_ln_tylor2(x)  # Aproximación de orden 2
y_taylor_3 = f_ln_tylor0(x) + df_ln_tylor1(x) + d2f_ln_tylor2(x) + d3f_ln_tylor3(x)  # Aproximación de orden 3
y_taylor_4 = f_ln_tylor0(x) + df_ln_tylor1(x) + d2f_ln_tylor2(x) + d3f_ln_tylor3(x) + d4f_ln_tylor4(x)  # Aproximación de orden 4
plt.figure(figsize=(10, 6))
plt.plot(x, y_funcion, label='f(x) = ln(x)', color='blue')
plt.plot(x, y_taylor_1, label='Taylor orden 1', linestyle='--')
plt.plot(x, y_taylor_2, label='Taylor orden 2', linestyle='--')
plt.plot(x, y_taylor_3, label='Taylor orden 3', linestyle='--')
plt.plot(x, y_taylor_4, label='Taylor orden 4', linestyle='--')
plt.title('Función ln(x) y Aproximaciones de Taylor')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.xlim(0, 3)
plt.ylim(-2, 2)
plt.grid(True)
plt.show()
```

## Parte dos del taller- Polinomio de Lagrange
### Encuentre el polinomio de Lagrange para los siguientes datos y grafique:


```python
## (0,0),(30,0.5),(60,3√2),(90,1)

# El polinomio de lagrande para los puntos dados es:
def lagrange(x):
    L0 = ((x - 30) * (x - 60) * (x - 90)) / ((0 - 30) * (0 - 60) * (0 - 90))
    L1 = ((x - 0) * (x - 60) * (x - 90)) / ((30 - 0) * (30 - 60) * (30 - 90))
    L2 = ((x - 0) * (x - 30) * (x - 90)) / ((60 - 0) * (60 - 30) * (60 - 90))
    L3 = ((x - 0) * (x - 30) * (x - 60)) / ((90 - 0) * (90 - 30) * (90 - 60))
    
    P = L0*0 + L1*0.5 + L2*(3*np.sqrt(2)) + L3*1
    return P

## Graficar la función de Lagrange y los puntos dados
x = np.linspace(-10, 100, 1000)
y_lagrange = lagrange(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y_lagrange, label='Polinomio de Lagrange', color='blue')
plt.scatter([0, 1, 30, 60, 90], [0, -0.14,0.5, 3*np.sqrt(2), 1], color='red', label='Puntos dados')# Se añade un punto de prueba para 
## Observar como el polinomio de lagrange se ajusta a los puntos dados. 
plt.title('Polinomio de Lagrange para los puntos dados')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.xlim(-10, 100)
plt.ylim(-1, 5) 
plt.grid(True)
plt.show()
```

```python
## (1,1),(2,2),(3,2)

def lagrange_2(x):
    L0 = ((x - 2) * (x - 3)) / ((1 - 2) * (1 - 3))
    L1 = ((x - 1) * (x - 3)) / ((2 - 1) * (2 - 3))
    L2 = ((x - 1) * (x - 2)) / ((3 - 1) * (3 - 2))
    
    P = L0*1 + L1*2 + L2*2
    return P
## Graficar la función de Lagrange y los puntos dados
x = np.linspace(0, 4, 1000) 
y_lagrange_2 = lagrange_2(x)    
plt.figure(figsize=(10, 6))
plt.plot(x, y_lagrange_2, label='Polinomio de Lagrange', color='blue')
plt.scatter([1, 2, 3], [1, 2, 2], color='red', label='Puntos dados')
plt.title('Polinomio de Lagrange para los puntos dados')    
plt.xlabel('x')
plt.ylabel('P(x)')  
plt.axhline(0, color='black', lw=0.5, ls='--')  
plt.axvline(0, color='black', lw=0.5, ls='--')  
plt.xlim(0, 4)
plt.ylim(0, 3)
plt.grid(True)  
plt.show()
```

```python
## (−2,5),(1,7),(3,11),(7,34)
def lagrange_3(x):
    L0 = ((x - 1) * (x - 3) * (x - 7)) / ((-2 - 1) * (-2 - 3) * (-2 - 7))
    L1 = ((x + 2) * (x - 3) * (x - 7)) / ((1 + 2) * (1 - 3) * (1 - 7))
    L2 = ((x + 2) * (x - 1) * (x - 7)) / ((3 + 2) * (3 - 1) * (3 - 7))
    L3 = ((x + 2) * (x - 1) * (x - 3)) / ((7 + 2) * (7 - 1) * (7 - 3))
    
    P = L0*5 + L1*7 + L2*11 + L3*34
    return P
## Graficar la función de Lagrange y los puntos dados
x = np.linspace(-2, 7, 200)
y_lagrange_3 = lagrange_3(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y_lagrange_3, label='Polinomio de Lagrange', color='blue')
plt.scatter([-2, 1, 3, 7], [5, 7, 11, 34], color='red', label='Puntos dados')
plt.title('Polinomio de Lagrange para los puntos dados')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.xlim(-3, 10) 
plt.ylim(0, 40)
plt.grid(True)
plt.show()
print(lagrange_3(3))
```
