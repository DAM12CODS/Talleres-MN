```python
def secant_method(f, x0, x1, tol=1e-4, max_iter=100):    
    x_prev = x0
    x_curr = x1
    iter_count = 0

    while abs(f(x_curr)) > tol and iter_count < max_iter:
        # Calculate the next approximation using the secant method formula
        x_next = x_curr - f(x_curr) * (x_curr - x_prev) / (f(x_curr) - f(x_prev))
        print("valor", x_next)
        # Update variables for the next iteration
        x_prev = x_curr
        x_curr = x_next
        iter_count += 1

    return x_curr, iter_count

print ("Secant method:")
x_root, iterations = secant_method(f, 0.5, 1.0)
print(f"Root: {x_root}, Iterations: {iterations}")
```
