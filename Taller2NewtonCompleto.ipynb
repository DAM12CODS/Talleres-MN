#   X_0 = estable

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# FUNCIÓN Y DERIVADA
# ==========================================

def f(x):
    return x**3 - 3 * x**2 + x - 1

def df(x):
    return 3 * x**2 - 6 * x + 1

# ==========================================
# MÉTODO DE NEWTON
# ==========================================

x0 = -8
tol = 1e-8
max_iter = 50

iteraciones = [x0]

x = x0
error = 1
i = 0

while error > tol and i < max_iter:

    fx = f(x)
    dfx = df(x)

    # Evitar división por cero
    if abs(dfx) < 1e-12:
        print("Derivada cercana a cero")
        break

    # Fórmula de Newton
    xn = x - fx/dfx

    # Error
    error = abs(xn - x)

    # Guardar iteración
    iteraciones.append(xn)

    # Actualizar valor
    x = xn

    i += 1

# ==========================================
# RESULTADOS
# ==========================================

print("Raíz aproximada:", x)

for j, val in enumerate(iteraciones):
    print(f"i = {j}, x_i = {val}")

# ==========================================
# GRÁFICA x_i vs i
# ==========================================

indices = np.arange(len(iteraciones))

plt.figure(figsize=(10,6))

plt.plot(indices,
         iteraciones,
         marker='o',
         linestyle='-',
         color='blue')

plt.title("Convergencia del Método de Newton")
plt.xlabel("Iteración i")
plt.ylabel("x_i")

plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# FUNCIÓN Y DERIVADA
# ==========================================

def fun(x):
    return x**3 - 3*x**2 + x - 1

def derfun(x):
    return 3*x**2 - 6*x + 1

# ==========================================
# NEWTON
# ==========================================

x0 = 100
max_iter = 20
tol = 1e-8

iteraciones = []

x = x0

for i in range(max_iter):

    iteraciones.append(x)

    fx = fun(x)
    dfx = derfun(x)

    if abs(dfx) < 1e-12:
        break

    x_new = x - fx/dfx

    if abs(x_new - x) < tol:
        iteraciones.append(x_new)
        break

    x = x_new

# ==========================================
# PUNTOS (x_i, f(x_i))
# ==========================================

xi = np.array(iteraciones)
yi = fun(xi)

# ==========================================
# GRÁFICA
# ==========================================

x_vals = np.linspace(-2, 5, 5000)

plt.figure(figsize=(10,6))

# función
plt.plot(x_vals, fun(x_vals), label="f(x)", color="blue")
plt.axhline(0, color="black")

# ==========================================
# LÍNEA QUE UNE LOS PUNTOS
# ==========================================

plt.plot(xi, yi, '-o', color="blue", label="Iteraciones Newton")

# ==========================================
# SOLO 5 TANGENTES
# ==========================================

for k in range(min(5, len(xi))):

    xk = xi[k]
    yk = yi[k]
    mk = derfun(xk)

    xt = np.linspace(xk - 1, xk + 1, 100)
    yt = yk + mk*(xt - xk)

    plt.plot(xt, yt, '--', alpha=0.5)
    #plt.plot(xt, yt, '--', alpha=0.5, linewidth=3)
# ==========================================
# FORMATO
# ==========================================

plt.title("Newton: línea de iteraciones + tangentes")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

plt.show()
```
