```python
import numpy as np
import matplotlib.pyplot as plt

# Crear lista de tokens
tokens = list(range(1_000_000))

# Lista para almacenar precios
precio = []

for t in tokens:
    # Primera recta: de 0 a 300k
    if t <= 300_000:
        y = 0.00005 * t  # pendiente menor
    else:
        # Segunda recta: continúa desde 300k con mayor pendiente
        y_300k = 0.00005 * 300_000
        y = y_300k + 0.00015 * (t - 300_000)

    precio.append(y)

# Graficar
plt.figure(figsize=(10, 5))
plt.plot(tokens, precio)

# Línea vertical opcional para marcar el cambio
plt.axvline(x=300_000, linestyle='--')

plt.xlabel("Tokens")
plt.ylabel("Precio")
plt.title("Precio vs Tokens")
plt.grid(True)

plt.show()
```
