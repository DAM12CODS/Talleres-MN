1. Subir notebook con la gráfica de precio vs tokens para el modelo GPT 5.5 y diferenciar en los casos de short context y long context. 


```python
import numpy as np 
```

```python

import matplotlib.pyplot as plt


tokens = list(range(1_000_000))
precio = []

for t in tokens:
  
    if t <= 272_000:
        y = 0.00005 * t  # a manera de artilugio se uso una pendiente de 0.00005 para la primera recta
    else:
        y_272k = 0.00005 * 272_000
        y = y_272k + 0.00015 * (t - 272_000) #a manera de artilugio se uso una pendiente de 0.00015 para la segunda recta, 
        #y se sumo el valor en el punto de cambio para mantener la continuidad    


    precio.append(y)


plt.figure(figsize=(10, 5))
plt.plot(tokens, precio)

# Linea vertical opcional para marcar el cambio
plt.axvline(x=272_000, linestyle='--')

plt.xlabel("Tokens")
plt.ylabel("Precio")
plt.title("Precio vs Tokens")
plt.grid(True)

plt.show()
```

2. Encontrar todas las raíces de la ecuación f(x)=1/4(x3+3x2−6x−8)
 usando el método de la bisección. Grafique. Modifique el número de iteraciones a 5. Pruebe los siguientes rangos de búsqueda ([-5 3], [-4.7 2.5], [-4.5 2.7], [-3 -2]). 


```python

# funcion que meayuda a usarla para el metodo de biseccion

def f(x):
    return (x**3 + 3*x**2 - 6*x - 8) / 4


# Método de bisección usando el pseudocódigo
def biseccion(a, b, tolerancia=0.0001):

    i = 1
    FA = f(a)

    print(f"\nIntervalo inicial: [{a}, {b}]")

    while i <= 5:

        # Paso 3
        p = a + (b - a) / 2
        FP = f(p)

        print(
            f"Iteración {i}: "
            f"a = {a:.6f}, "
            f"b = {b:.6f}, "
            f"p = {p:.6f}, "
            f"f(p) = {FP:.6f}"
        )

        # Paso 4
        if FP == 0 or (b - a) / 2 < tolerancia:
            print(f"\nRaíz aproximada encontrada: {p:.6f}")
            return p

        # Paso 5
        i = i + 1

        # Paso 6
        if FA * FP > 0:
            a = p
            FA = FP
        else:
            b = p

    # Paso 7
    print("\nEl método fracasó después de 5 iteraciones")

rangos = [
    (-5, 3),
    (-4.7, 2.5),
    (-4.5, 2.7),
    (-3, -2)
]

for a, b in rangos:
    biseccion(a, b)


x = np.linspace(-6, 4, 1000)
y = f(x)

plt.figure(figsize=(10,5))

plt.plot(x, y, label='f(x)')


plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Método de Bisección")
plt.grid(True)

plt.show()
```
